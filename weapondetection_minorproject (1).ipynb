```python
import os

from ultralytics import YOLO

model = YOLO("yolov8m.pt")
```

```python
results = model.train(data=os.path.join(ROOT_DIR, "/content/gdrive/MyDrive/yolov8session/weapon_detection.v5i.yolov8/data.yaml"), epochs=20)
```

```python
import os
from ultralytics import YOLO
from PIL import Image
import cv2
import numpy as np
from matplotlib import pyplot as plt
predictions = model.predict('/content/gdrive/MyDrive/yolov8session/weapon_detection.v5i.yolov8/test/images/DefenseKnifeAttack0202_jpg.rf.1d79baced85a47665110fbeef9b30da4.jpg')

# Load the image using PIL
image_path = '/content/gdrive/MyDrive/yolov8session/weapon_detection.v5i.yolov8/test/images/DefenseKnifeAttack0202_jpg.rf.1d79baced85a47665110fbeef9b30da4.jpg'
image = Image.open(image_path)
image_np = np.array(image)
```

```python
for prediction in predictions:
    boxes = prediction.boxes.xyxy.cpu().numpy()  # x1, y1, x2, y2
    scores = prediction.boxes.conf.cpu().numpy()
    class_ids = prediction.boxes.cls.cpu().numpy()

    for box, score, class_id in zip(boxes, scores, class_ids):
        x1, y1, x2, y2 = map(int, box)

        # Draw the bounding box
        cv2.rectangle(image_np, (x1, y1), (x2, y2), (0, 255, 0), 2)

        # Display class and confidence
        label = f'{model.names[int(class_id)]}: {score:.2f}'
        cv2.putText(image_np, label, (x1, y1 - 10), cv2.FONT_HERSHEY_SIMPLEX, 0.5, (0, 255, 0), 2)

# Convert the image back to RGB (since OpenCV uses BGR by default)
image_np_rgb = cv2.cvtColor(image_np, cv2.COLOR_BGR2RGB)

# Display the image with bounding boxes
plt.figure(figsize=(12, 8))
plt.imshow(image_np_rgb)
```
